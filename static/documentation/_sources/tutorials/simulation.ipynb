```python
class BolusOneCmtModel(pk.IvOneCmtLinear):
    def __init__(self) -> None:
        super().__init__()

        # self.cl = column('CL')
        # self.v = column('V')

        self.one = theta(1)
        self.eta = omega(1)
        self.eps = sigma(1)

    def pred(self):
        cl = 0.5 * self.one
        v = 3 * self.one

        computed = self.trans2(cl=cl, v=v)
        y = computed.F
        
        return y
```

```python
class SimulationSetting(ParameterSet):
    def __init__(self) -> None:
        super().__init__()
        
        self.one = theta(0.5)
        self.eta = omega(1)
        self.eps = sigma(1)
```
